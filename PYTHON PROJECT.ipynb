```python
from tkinter import*
from translate import Translator
from tkinter import ttk

# Create a GUI window
root = Tk()

# Set the background colour of GUI window
root.configure(background = '#20325B')

# Set the configuration of GUI window (WidthxHeight)
root.geometry("450x300")

# set the name of tkinter GUI window
root.title("Group XI:Python Project")

# defining varisbles for text translation
InputLanguageChoice = StringVar()
TranslateLanguageChoice = StringVar()
OutputVar = StringVar()
TextVar = StringVar()
InputLanguageChoice.set('Hindi')
TranslateLanguageChoice.set('English')

# defining the widgets
headlabel = Label(root, text = 'Text to Text Translator: Hindi to English',fg = 'black', bg = "#ffffcc")
label1 = Label(root, text = " Hindi Text ",fg = 'black', bg = '#ffffcc')
label2 = Label(root, text = " English Text",fg = 'black', bg = '#ffffcc')
text1_area = Text(root, height = 5, width = 25, font = "lucida 13")
text1_area = Entry(root,textvariable=TextVar)

text2_area = Text(root, height = 5, width = 25, font = "lucida 13")
text2_area = Entry(root,textvariable=OutputVar)



# grid method is used for placing the widgets at respective positionsin table like structure .
headlabel.grid(row = 0, column = 1)

label1.grid(row = 1, column = 0, padx = 10, pady = 10)
label2.grid(row = 3, column = 0, padx = 10, pady = 10)


text1_area.grid(row = 1, column = 1, padx = 10, pady = 10)
text2_area.grid(row = 3, column = 1, padx = 10, pady = 10)


#function to clear the input and output field
def clearAll() :
    text1_area.delete(0, END)
    text2_area.delete(0, END)
    
#function to translate
def Translate():
    translator = Translator(from_lang= InputLanguageChoice.get(),to_lang=TranslateLanguageChoice.get())
    Translation = translator.translate(TextVar.get())
    OutputVar.set(Translation)

# button for translation
button1 = Button(root, text = "Translate to English",bg = "#ffffcc", fg = "black", command =Translate)
button1.grid(row = 2, column = 1)

# Created a Clear Button and attached with clearAll function
button2 = Button(root, text = "Clear", bg = "#ffffcc",fg = "black", command = clearAll)
button2.grid(row = 5, column = 1)


root.mainloop()
```
